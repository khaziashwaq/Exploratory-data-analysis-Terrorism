```python
dt = dt[['iyear','imonth','iday','country_txt','provstate','region_txt','city','latitude','longitude',"attacktype1_txt",'nkill',
               'nwound','targtype1_txt','gname','weaptype1_txt']]
```

```python
dt.head()
```

```python
dt = dt.rename(columns= {'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State','region_txt':'Region','city':'City','attacktype1_txt':'Attacktype','nkill':'Kill','nwound':'Wound','targtype1_txt':'Targettype','weaptype1_txt':'Weapontype','gname':'Groupname'})
```

```python
dt.head()
```

```python
dt.isnull().sum()
```

```python
dt['Kill'] = dt['Kill'].fillna(0)
dt['Wound'] = dt['Wound'].fillna(0)
```

```python
dt.describe()
```

```python
year = dt['Year'].unique()
year_count = dt['Year'].value_counts(dropna=False).sort_index()
plt.figure(figsize=(20,10))
sns.barplot(x=year, y=year_count)
plt.xticks(rotation=60)
plt.xlabel('Year',fontsize=20)
plt.ylabel('Number of attacks',fontsize=20)
plt.title('Attacks vs Year',fontsize=20)
plt.show()
```

```python
df = dt[['Kill','Year']].groupby(['Year']).sum()
df.plot(kind='bar',figsize=(20,10))
plt.title('Death due to terrorism',fontsize=20)
plt.xlabel('Year',fontsize=20)
plt.ylabel('Number of Kills',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(dt['Country'].value_counts()[:20].index,dt['Country'].value_counts()[:20].values,palette='coolwarm_r')
plt.title('Top Countries affected by terrorism', fontsize=20)
sns.color_palette("rocket")
plt.xticks(rotation = 90)
plt.show()
```

```python
dt_killed = dt[['Kill','Country']].groupby(['Country']).sum().sort_values(by='Kill',ascending=False).head(20)
dt_killed.plot(kind='bar', figsize=(20,10))
plt.title('Total number of Citizens killed',fontsize=20)
plt.xlabel('Country',fontsize=20)
plt.ylabel('Number of people',fontsize=20)
plt.show()
```

```python
dt_killed = dt[['Wound','Country']].groupby(['Country']).sum().sort_values(by='Wound',ascending=False).head(20)
dt_killed.plot(kind='bar',figsize=(20,10))
plt.title('Total number of Citizens wounded',fontsize=20)
plt.xlabel('Country',fontsize=20)
plt.ylabel('Number of people',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(dt['City'].value_counts()[:20].index,dt['City'].value_counts()[:20].values,palette='PuRd_r')
plt.title('Top Cities affected by terrorism', fontsize=20)
plt.xticks(rotation = 90)
plt.show()
```

```python
dt1_killed = dt[['Kill','City']].groupby(['City']).sum().sort_values(by='Kill',ascending=False).head(20)
dt1_killed.plot(kind='bar',figsize=(20,10))
plt.title('Total number of Citizens got killed',fontsize=20)
plt.xlabel('City',fontsize=20)
plt.ylabel('Number of people',fontsize=20)
plt.show()
```

```python
dt_killed = dt[['Wound','City']].groupby(['City']).sum().sort_values(by='Wound',ascending=False).head(20)
dt_killed.plot(kind='bar',figsize=(20,10))
plt.title('Total number of people got wounded',fontsize=20)
plt.xlabel('City',fontsize=20)
plt.ylabel('Number of people',fontsize=20)
plt.show()
```

```python
dt1_killed = dt[['Kill','Region']].groupby(['Region']).sum().sort_values(by='Kill',ascending=False).head(20)
dt1_killed.plot(kind='bar', figsize=(20,10))
plt.title('Total number of People got killed',fontsize=20)
plt.xlabel('Region',fontsize=20)
plt.ylabel('Number of people',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(dt['Region'].value_counts()[:20].index,dt['Region'].value_counts()[:20].values,palette='YlOrRd_r')
plt.title('Top Regions affected by terrorism', fontsize=20)
sns.color_palette("rocket")
plt.xticks(rotation = 90)
plt.show()
```

```python
attac_type = dt[['Kill','Attacktype']].groupby('Attacktype').sum()
attac_type.plot(kind='bar',color='crimson',figsize=(20,10))
plt.title('Attack type vs Number of attacks')
plt.show()
```

```python
dt['Attacktype'].value_counts().plot(kind='bar',color='cyan',figsize=(20,10))
plt.title('Number of attacks vs Attack Type',fontsize=20)
plt.show()
```

```python
weapon_type = dt[['Kill','Weapontype']].groupby('Weapontype').sum()
weapon_type.plot(kind='bar',color='crimson',figsize=(20,10))
plt.title('Attack type vs Number of attacks')
plt.show()
```

```python
dt['Weapontype'].value_counts().plot(kind='bar',color='cyan',figsize=(20,10))
plt.title('Number of attacks vs Attack Type',fontsize=20)
plt.show()
```

```python
dt['Groupname'].value_counts().drop('Unknown').head(20).plot(kind='bar',color='Orange',figsize=(20,10))
plt.show()
```

```python
group_ = dt[['Kill','Groupname']].groupby('Groupname').sum().drop('Unknown').sort_values('Kill',ascending=False).head(20)
group_.plot(kind='bar',figsize=(20,10))
plt.show()
```

```python
dt['Groupname'].value_counts().drop('Unknown').head(20).plot(kind='pie',autopct='%1.1f%%',figsize=(40,40),startangle=180,fontsize='50')
plt.title("Groupname attacks in total",fontsize=70)
plt.show()
```

```python
dt['Attacktype'].value_counts().plot(kind='pie',figsize=(20,20),autopct='%1.1f%%',fontsize='30')
plt.title("Attacktype attacks in total",fontsize=40)
plt.show()
```

```python
dt['Weapontype'].value_counts().plot(kind='pie',figsize=(20,20),autopct='%1.1f%%',fontsize='30')
plt.title("Weapontypes attack in total",fontsize=40)
plt.show()
```
